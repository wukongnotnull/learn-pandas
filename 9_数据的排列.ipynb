```python
import pandas as pd
```

```python
df = pd.read_csv('./datas/weather.csv')
df
```

```python
# sort_values(ascending='默认升序True/False降序') 升序和降序 
df['相对湿度(%)'].sort_values(ascending=True)
```

```python
df['相对湿度(%)'].sort_values(ascending=False)
```

```python
# 单列排序
df.sort_values(by='相对湿度(%)',ascending=False)
```

```python
# 多列排序
df.sort_values(by=['相对湿度(%)','累积雨量(mm)'],ascending=[True,False]).tail(20) 
```
