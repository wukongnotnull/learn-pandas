# 读取 weather 文件


```python
import pandas as pd
```

```python
df = pd.read_csv('./datas/weather.csv')
df.head()
```

```python
# 不使用默认行索引，使用日期作为行索引
df.set_index('日期',inplace=True)
df
```

```python
df.index
```

```python
df.columns
```

```python
# str 类中的 replace 函数 替换的作用
df.loc[:,'气温(度)'] = df['气温(度)'].str.replace('℃','')
```

```python
df
```

```python
df['气温(度)']
```

```python
# 自定义行索引和列索引的时候，不存在左闭右开
df.loc['2015-01-02':'2015-01-05','城市':'观测站']
```

```python
# 若是使用默认的索引，就遵守左闭右开
df.iloc[1:5,0:3]
```

```python
# 条件表达式 < 的方式查询
df.loc[df['气温(度)'].astype(float) < 10,:]
```

```python
# 条件表达式 < 的方式查询
df.loc[(df['气温(度)'].astype(float) < 10) & (df['相对湿度(%)'].astype(float) > 0),:]
```

## 通过函数的方式进行查询

```python
# 函数的方式定义查询条件
def select(x):
    return (x['气温(度)'].astype(float) < 10) & (x['相对湿度(%)'].astype(float) > 0)

# 第一参数为函数名
df.loc[select,:]
```

## 使用匿名函数作为查询条件 
lambda 关键字定义匿名函数

```python
df.loc[lambda x: (x['气温(度)'].astype(float) < 10) & (x['相对湿度(%)'].astype(float) > 0) ,:]
```
