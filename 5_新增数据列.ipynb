```python
# 导入 pandas 第三方包
import pandas as pd
```

```python
df = pd.read_csv('./datas/weather.csv')
df
```

# 直接赋值法

```python
df.loc[:,'气温度值'] = df['气温(度)'].str.replace('℃','').astype(float)
df
```

```python
df.loc[:,'温差'] = df['气温度值'] - 10
df
```

## df.apply() 方法
axis 轴，为1 表示列，追加新列

         为 0 ，表示行，追加新行

```python
# 自定义函数 x 表示：DataFrame 对象
def get_temperature_type (x):
    if x['气温度值'] > 20:
        return '高温'
    elif x['气温度值'] > 5:
        return '常温'
    else:
        return '低温'

df.loc[:,'温度类型'] = df.apply(get_temperature_type,axis = 1)
df.head(20)
```

```python
# 统计三种温度类型分别是多少个 value 值 counts 计数
df['温度类型'].value_counts()
```
