```python
# dropna() drop 删除 na 等于 null 等于 NaN，就是空值 缺失值
# axis ='' ,1或者columns 表示按列删除； 0或者  index 表示按照行删除
# how : all值全为NaN ；any 值任何为NaN
#df.dropna(axis='按行还是列操作',how ='all/any'，inplace=‘True/False’)
```

```python
df3 =df.dropna(axis='columns',how='all',inplace =False)
df3
```

```python
df3.dropna(axis='index',how='all',inplace=True)
```

```python
df3
```

```python
# fillna(method='ffill/bfill')
# ffill forward fill 向前填充 
# bfill backward fill 向后填充
```

```python
df3['姓名'].fillna(method='ffill')
```

```python
df3.loc[:,'姓名'] = df3['姓名'].fillna(method='ffill')
df3
```

```python
# 传统方法
df3.loc[:,'分数'] = df3['分数'].fillna(0)
df3
```

```python
# 使用字典方式
# df3.fillna({"分数":0})
```

```python
# 数据清洗后，可以输出到一个新的excel 表中 
df3.to_excel('./datas/student3.xlsx',index=False)
```
